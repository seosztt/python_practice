```python
import numpy as np

x=np.arange(-5,5,0.1)
y1=x**2 -2
y2=20*np.cos(x)**2

plt.figure(1)
plt.plot(x,y1)

plt.figure(2)
plt.plot(x,y2)

plt.figure(1)
plt.plot(x,y2)

plt.figure(2)
plt.clf()
plt.plot(x,y1)

plt.show()
```

```python
import numpy as np
x=np.arange(0,10,0.1)
y1=0.3*(x-5)**2+1
y2= -1.5*x+3
y3=np.sin(x)**2
y4=10*np.exp(-x)+1

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

```python
import numpy as np
x=np.arange(0,5,1)
y1=x
y2=x+1
y3=x+2
y4=x+3
```

```python
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

```python
plt.plot(x,y1,'m',x,y2,'y',x,y3,'k',x,y4,'c')
plt.show()
```

```python
plt.plot(x,y1,'-',x,y2,'--',x,y3,':',x,y4,'-.')
plt.show()
```

```python
plt.plot(x,y1,'o',x,y2,'^',x,y3,'s',x,y4,'d')
plt.show()
```

```python
plt.plot(x,y1,'>--r',x,y2,'s-g',x,y3,'d:b',x,y4,'-.Xc')
plt.show()
```

```python
import numpy as np
x=np.arange(0,5,1)
y1=x
y2=x+1
y3=x+2
y4=x+3

plt.plot(x,y1, '>--r',x,y2,'s-g',x,y3,'d:b',x,y4,'-.Xc')
plt.legend(['data1','data2','data3', 'data4'])
plt.show()
```

```python
import numpy as np
x=np.arange(0,5,1)
y1=x
y2=x+1
y3=x+2
y4=x+3

plt.plot(x,y1, '>--r',x,y2,'s-g',x,y3,'d:b',x,y4,'-.Xc')
plt.legend(['data1','data2','data3', 'data4'], loc='lower right')
plt.show()
```
